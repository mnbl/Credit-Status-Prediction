```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

Reference Dataset: [Link](https://www.kaggle.com/datasets/bhavikjikadara/loan-status-prediction)

```python
df = pd.read_csv("./loan_data.csv")
df.columns = [col.lower() for col in df.columns]
df.head()
```

```python
df['applicant_employment'] = df.self_employed.apply(lambda x: "Self-Employed" if str(x).lower() == "yes" else "Salaried")
df['applicant_married'] = df['married'].apply(lambda x: True if str(x).lower() == "yes" else False)
df['loan_status'] = df['loan_status'].apply(lambda x: True if str(x).lower() == "y" else False)
# df["applicantincome"] = df["applicantincome"].apply(lambda x: int(x)*12)
# df["coapplicantincome"] = df["coapplicantincome"].apply(lambda x: int(x)*12)
# df["loanamount"] = df["loanamount"].apply(lambda x: x*1000)
df.drop(["self_employed", "credit_history", "married"], axis = 1, inplace = True)
```

```python
df['co_applicant'] = np.random.choice([True, False], size=len(df))
df['co_applicant_employment'] = df.apply(lambda row: np.random.choice(["Self-Employed", "Salaried"]) if row['co_applicant'] else "Not Applicable", axis=1)
df['co_applicant_employment_history'] = df['co_applicant_employment'].apply(lambda x: np.random.randint(0, 49) if x != "Not Applicable" else 0)
df['co_applicant_credit_score'] = np.where(df['loan_status'] == True, np.random.randint(600, 901, size=len(df)), np.random.randint(400, 700, size=len(df)))
df["co_applicant_credit_score"] = df.apply(lambda row: row["co_applicant_credit_score"] if row["co_applicant"] else 650, axis=1)
```

```python
df['co_applicant_credit_default'] = np.where(df['loan_status'] == True, np.random.choice(['True', 'False'], p=[0.2, 0.8], size=len(df)), np.random.choice(['True', 'False'], p=[0.8, 0.2], size=len(df)))
df["co_applicant_credit_default"] = df.apply(lambda row: row["co_applicant_credit_default"] if row["co_applicant"] else False, axis=1)
```

```python
df['applicant_employment_history'] = df["co_applicant_employment_history"].apply(lambda x: np.random.randint(x, 61))
df['applicant_credit_score'] = np.where(((df['loan_status'] == True) | (df['co_applicant_credit_score'] > 650)),
                                        np.random.randint(651, 901, size=len(df)),
                                        np.random.randint(300, 651, size=len(df)))
df['applicant_credit_default'] = np.where(df['loan_status'] == True, np.random.choice(['True', 'False'], p=[0.2, 0.8], size=len(df)), np.random.choice(['True', 'False'], p=[0.8, 0.2], size=len(df)))
df['applicant_credit_history'] = df["applicant_employment_history"].apply(lambda x: np.random.randint(x, 61))
```
